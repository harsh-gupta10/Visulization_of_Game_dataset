```python
import pandas as pd
```

```python
df = pd.read_csv ('../Data/video_games_sales.csv').dropna() 
#dropping the na values at the import stage itself
df.head()
```

```python
df = df[['Name','Global_Sales','Platform']]
```

```python
df = df.rename(columns={'Name': 'id', 'Platform': 'parent','Global_Sales':'value'})
```

```python
result = df.groupby('parent').apply(lambda x: x.nlargest(6, 'value')).reset_index(drop=True)
```

```python
result
```

```python
result.to_csv('../Data/Sunburst.csv', index=False)
```
