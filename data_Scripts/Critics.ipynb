```python
import pandas as pd
```

```python
df = pd.read_csv('../Data/video_games_sales.csv')
df2 = df[['Name','User_Score','Critic_Score']]
```

```python
df2['User_Score'] = pd.to_numeric(df2['User_Score'], errors='coerce')
df2['Critic_Score'] = pd.to_numeric(df2['Critic_Score'], errors='coerce')

# Remove rows with NaN values in 'User_Score' and 'Critic_Score'
df2 = df2[pd.notnull(df2['User_Score']) & pd.notnull(df2['Critic_Score'])]
# df2.to_csv('../Data/critics.csv', index=False)

```

```python
min_score = df2['User_Score'].min()
max_score = df2['User_Score'].max()
df2['Normalized_User_Score'] = (df2['User_Score'] - min_score) / (max_score - min_score)


min_score = df2['Critic_Score'].min()
max_score = df2['Critic_Score'].max()
df2['Normalized_Critic_Score'] = (df2['Critic_Score'] - min_score) / (max_score - min_score)

df2.to_csv('../Data/critics.csv', index=False)
df2
```
