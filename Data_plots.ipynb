# Sales Trends of Top 10 Publishers
**Important note:** Sales are grouped by release year, not by the sale date since we do not have that data 

```python
top_publishers=10
vg=vg_data.groupby(['Publisher']).agg({"Global_Sales":sum})
vg.reset_index(inplace=False)
vg=vg.sort_values(by='Global_Sales', ascending=False)

top_publishers=list(vg.head(10).index)

top_publishers
```

```python
# Grouped Data
vg=vg_data[vg_data['Publisher'].isin(top_publishers)].groupby(['Year','Publisher']).agg({'Global_Sales':sum})
vg=vg.sort_values(by='Year', ascending=True)
vg.reset_index(inplace=True)
vg
```

Let us now plot Sales trends of the Top 10 publishing companies

```python
fig=px.line(vg, x='Year', y='Global_Sales', color='Publisher')
fig.update_layout(title_text='Top 10 Publishers Sales Trends')
fig.update_xaxes(title_text="Release Year")
fig.update_yaxes(title_text="Global_sales")

fig.show()
```

This is the multi-line chart that confirms insights seen from point 4

# Video Game Sales by Genre by Decades

```python
# Cook the data
vg_1980s=vg_data.query("Year>=1980 and Year <2000")[['Genre','Global_Sales']].groupby(['Genre']).sum()
vg_1980s.reset_index(inplace=True)

vg_1990s=vg_data.query('Year>=1990 and Year<2000')[['Genre','Global_Sales']].groupby(['Genre']).sum()
vg_1990s.reset_index(inplace=True)

vg_2000s=vg_data.query('Year>=2000 and Year<2010')[['Genre','Global_Sales']].groupby(['Genre']).sum()
vg_2000s.reset_index(inplace=True)

vg_2010s=vg_data.query('Year>=2010')[['Genre','Global_Sales']].groupby('Genre').sum()
vg_2010s.reset_index(inplace=True)
```

```python
# Create subplots: use 'domain' type fpr Pie subplot
labels=['USA','China','European Union','Russian Federation','bRAZIL','India',' Rest of World']
fig=make_subplots(rows=2, cols=2, specs=[[{'type':'domain'},{'type':'domain'}],[{'type':'domain'}, {'type':'domain'}]])

fig.add_trace(go.Pie(labels=vg_1980s['Genre'], values=vg_1980s['Global_Sales'], name="1980s", title="1980s", hole=.3), 1, 1)
fig.add_trace(go.Pie(labels=vg_1990s['Genre'], values=vg_1990s['Global_Sales'], name="1990s", title="1990s", hole=.3), 1, 2)
fig.add_trace(go.Pie(labels=vg_2000s['Genre'], values=vg_2000s['Global_Sales'], name="2000s", title="2000s", hole=.3), 2, 1)
fig.add_trace(go.Pie(labels=vg_2010s['Genre'], values=vg_2010s['Global_Sales'], name="2010s", title="2010s", hole=.3), 2, 2)
fig.update_layout(height=800, title_text="Video Game Sales by Genre by Decades")
fig.show()
```

## Insights
* in the 80s, **Platform** games were most played, with 20% now, in th 2010s it has decreased to 4.79%.



* Instead, the **Action** genre is now the most famous with 26.7%.



*  The 90s  marked the rise of **RPG** games  and the decline of the Platform games.



* The 2000s was when **esports** games had a boom reppn 17.6% of the gaming market
